```python
import json
import os
from google.colab import drive
import ipywidgets as widgets
from IPython.display import display, HTML

# 1. חיבור ל-Google Drive
drive.mount('/content/drive')

# נתיב לקובץ בדרייב
PATH = '/content/drive/MyDrive/students.json'

def load_data():
    try:
        with open(PATH, 'r', encoding='utf-8') as f:
            return json.load(f)
    except FileNotFoundError:
        return []

def save_data(data):
    with open(PATH, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

# טעינה ראשונית
students_list = load_data()

# --- עיצוב מקצועי וקריא ---
display(HTML("""
<style>
    .custom-form {
        background-color: #000000; /* Changed to black */
        color: #ffffff; /* Changed text color to white */
        border: 1px solid #333333; /* Darker border for contrast */
        border-radius: 4px;
        padding: 20px;
        font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    }
    .main-header {
        color: #ffffff; /* Header text color to white */
        border-bottom: 2px solid #555555; /* Lighter border for contrast */
        margin-bottom: 20px;
        padding-bottom: 10px;
    }
    .widget-label { font-weight: 600 !important; color: #ffffff !important; } /* Ensure widget labels are white */
    .success-msg {
        background-color: #2e7d32; /* Darker green for success on black */
        color: #ffffff;
        padding: 10px;
        border-radius: 4px;
        border: 1px solid #66bb6a;
        margin-top: 15px;
        font-weight: bold;
    }
    /* Adjusting text color for widgets to be visible on dark background */
    .widget-text input, .widget-textarea textarea {
        background-color: #333333 !important;
        color: #ffffff !important;
        border: 1px solid #555555 !important;
    }
    .widget-dropdown select {
        background-color: #333333 !important;
        color: #ffffff !important;
        border: 1px solid #555555 !important;
    }
</style>
"""))

# --- 2. יצירת ווידג'טים ---
style = {'description_width': '130px'}
layout = widgets.Layout(width='420px', margin='5px 0')

dropdown_options = [(f"{s['first_name']} {s['last_name']}", i) for i, s in enumerate(students_list)]

dropdown = widgets.Dropdown(options=dropdown_options, description='בחר סטודנט:', style=style, layout=layout)
first_name_w = widgets.Text(description='שם פרטי:', disabled=True, style=style, layout=layout)
last_name_w = widgets.Text(description='שם משפחה:', disabled=True, style=style, layout=layout)
email_w = widgets.Text(description='מייל בראודה:', disabled=True, style=style, layout=layout)
courses_w = widgets.Textarea(description='קורסים:', disabled=True, style=style, layout=layout)
link_w = widgets.Text(description='קישור מעניין:', disabled=True, style=style, layout=layout)
fav_program_w = widgets.Text(description='תוכנית אהובה:', placeholder='הזן שם תוכנית...', style=style, layout=layout)

update_btn = widgets.Button(
    description='עדכן נתונים',
    button_style='primary',
    icon='save',
    layout=widgets.Layout(width='140px', height='35px', margin='20px 0 0 135px')
)

output = widgets.Output()

# --- 3. לוגיקה ---

def update_ui(index):
    if index is not None and index < len(students_list):
        s = students_list[index]
        first_name_w.value = s['first_name']
        last_name_w.value = s['last_name']
        email_w.value = s['email']
        courses_w.value = ", ".join(s['courses'])
        link_w.value = s['interesting_link']
        fav_program_w.value = s.get('favorite_program', '')

def on_dropdown_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        update_ui(change['new'])
        with output:
            output.clear_output()

def on_click_update(b):
    idx = dropdown.value
    students_list[idx]['favorite_program'] = fav_program_w.value
    save_data(students_list)

    with output:
        output.clear_output()
        display(HTML(f"<div class='success-msg'>✅ הנתונים עבור {first_name_w.value} נשמרו בהצלחה בדרייב.</div>"))

dropdown.observe(on_dropdown_change)
update_btn.on_click(on_click_update)

# --- 4. תצוגה ---
header = widgets.HTML("<div class='main-header'><h3>ניהול נתוני סטודנטים - Braude College</h3></div>")

form_container = widgets.VBox([
    header,
    dropdown,
    widgets.HTML("<div style='height: 10px;'></div>"),
    first_name_w, last_name_w, email_w, courses_w, link_w,
    widgets.HTML("<hr style='border: 0; border-top: 1px solid #eee; margin: 15px 0;'>"),
    fav_program_w,
    update_btn,
    output
], layout=widgets.Layout(padding='20px', width='480px'))

form_container.add_class('custom-form')
display(form_container)
update_ui(dropdown.value)
```
