```python
import requests
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
from ipywidgets import VBox, HBox

pd.set_option("display.max_columns", None)

url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)
data_df = pd.DataFrame(data['result']['records'])

# --- Widgets ---
# Manufacturer Dropdown
unique_manufacturers = sorted(data_df['tozeret_nm'].dropna().unique().tolist())
# Add a "Select manufacturer" option with a None value
manufacturer_options = [('בחר יצרן', None)] + [(m, m) for m in unique_manufacturers]
manufacturer_dropdown = widgets.Dropdown(
    options=manufacturer_options,
    value=None, # Initially nothing selected
    description='יצרן:', # Manufacturer in Hebrew
    disabled=False,
)

# Model Dropdown (initially empty or with 'Select model')
model_options = [('בחר דגם', None)]
model_dropdown = widgets.Dropdown(
    options=model_options,
    value=None, # Initially nothing selected
    description='דגם:', # Model in Hebrew
    disabled=False,
)

# Output area for results
results_output = widgets.Output()

# --- Functions ---

def display_current_selection_info():
    """
    Filters the data based on current dropdown selections and displays the info.
    If no manufacturer or no model is selected, clears the output.
    """
    with results_output:
        clear_output(wait=True) # Clear existing output in this widget

        selected_manufacturer = manufacturer_dropdown.value
        selected_model = model_dropdown.value

        # If no manufacturer or no model is selected, display nothing
        if selected_manufacturer is None or selected_model is None:
            return

        current_filtered_df = data_df.copy()

        # Filter by manufacturer if one is selected
        if selected_manufacturer is not None:
            current_filtered_df = current_filtered_df[current_filtered_df['tozeret_nm'] == selected_manufacturer]

        # Filter by model if one is selected
        if selected_model is not None:
            current_filtered_df = current_filtered_df[current_filtered_df['degem_nm'] == selected_model]

        # Display summary information only if there's data after filtering
        if not current_filtered_df.empty:
            print(f"סה\"כ רכבים: {len(current_filtered_df)} ({len(current_filtered_df.drop_duplicates(subset='mispar_rechev'))} רכבים ייחודיים)\n")

            kinuy_mishari_unique = current_filtered_df['kinuy_mishari'].dropna().unique()
            if len(kinuy_mishari_unique) > 0:
                print(f"כינויים מסחריים: {', '.join(kinuy_mishari_unique)}\n")
            else:
                print("אין כינויים מסחריים עבור בחירה זו.\n")

            ramat_gimur_unique = current_filtered_df['ramat_gimur'].dropna().unique()
            if len(ramat_gimur_unique) > 0:
                print(f"רמות גימור: {', '.join(ramat_gimur_unique)}\n")
            else:
                print("אין רמות גימור עבור בחירה זו.\n")
        else:
            print("אין נתונים עבור הבחירה הנוכחית.") # No data for current selection


def on_manufacturer_change(change):
    selected_manufacturer = change.new

    # Update model dropdown options
    if selected_manufacturer is None:
        model_dropdown.options = [('בחר דגם', None)]
        model_dropdown.value = None
    else:
        filtered_by_manufacturer = data_df[data_df['tozeret_nm'] == selected_manufacturer].copy()
        unique_models = sorted(filtered_by_manufacturer['degem_nm'].dropna().unique().tolist())
        model_dropdown.options = [('בחר דגם', None)] + [(m, m) for m in unique_models]
        model_dropdown.value = None # Reset model selection when manufacturer changes

    display_current_selection_info()


def on_model_change(change):
    # This function simply triggers the display update based on current selections
    display_current_selection_info()

# --- Link widgets to functions ---
manufacturer_dropdown.observe(on_manufacturer_change, names='value')
model_dropdown.observe(on_model_change, names='value')

# --- Initial Display ---
display(VBox([
    HBox([manufacturer_dropdown, model_dropdown]),
    results_output
]))

# Initial call to display info (which will be blank since manufacturer_dropdown.value is None)
display_current_selection_info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # Keep numpy as it might be used elsewhere, although not for the new chart logic

# Tab 1: Data Overview
tab1_content = widgets.Output()
with tab1_content:
    print("Data Overview:")
    display(data_df.describe())  # Summary statistics for the overall data

# Tab 2: Raw Data
tab2_content = widgets.Output()
with tab2_content:
    print("Raw Data:")
    display(data_df)  # Full data_df as requested

# Tab 3: Charts
tab3_content = widgets.Output()
with tab3_content:
    print("Charts: Number of Unique Cars Produced Each Year")
    plt.clf()  # Clear the current figure

    # Ensure 'shnat_yitzur' (production year) is numeric and handle missing values
    cars_per_year_df = data_df.dropna(subset=['shnat_yitzur']).copy()
    cars_per_year_df['shnat_yitzur'] = cars_per_year_df['shnat_yitzur'].astype(int)

    # Group by year and count unique vehicle IDs ('mispar_rechev')
    cars_count_by_year = cars_per_year_df.groupby('shnat_yitzur')['mispar_rechev'].nunique().reset_index()

    # Sort by year for better visualization
    cars_count_by_year = cars_count_by_year.sort_values('shnat_yitzur')

    # Create the bar chart
    plt.figure(figsize=(12, 6))
    sns.barplot(x='shnat_yitzur', y='mispar_rechev', data=cars_count_by_year, palette='viridis')
    plt.title('Number of Unique Cars Produced Each Year')
    plt.xlabel('Production Year')
    plt.ylabel('Number of Unique Cars')
    plt.xticks(rotation=45) # Rotate x-axis labels for readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout() # Adjust layout to prevent labels from being cut off
    plt.show() # Display the plot
    plt.close() # Close the figure to free memory
```

```python
# Create Tabs
tabs = widgets.Tab(children=[tab1_content, tab2_content,tab3_content])
tabs.set_title(0, 'Tab 1: Data Overview')
tabs.set_title(1, 'Tab 2: Raw Data')
tabs.set_title(2, 'Tab 3: Charts')



# Display Tabs
display(tabs)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Call once to configure Bokeh to display plots inline in the notebook.
output_notebook()
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = ["#%02x%02x%02x" % (r, g, 150) for r, g in zip(np.floor(50+2*x).astype(int), np.floor(30+2*y).astype(int))]

p = figure()
p.circle(x, y, radius=radii, fill_color=colors, fill_alpha=0.6, line_color=None)
show(p)

# --- הוספת ניתוח אחוזים לקונסולה ---
print('\n\n## ניתוח קוד - חלק יצירת צבעים לעומת חלקים אחרים')

# Get the current cell's code content
import inspect
import re

# Get the source code of the current cell dynamically
# This is a bit of a hack as direct cell content access can be tricky in Colab
# For simplicity, we will simulate reading the relevant part of the code.
# In a real dynamic scenario, one would need to access the cell content directly.

# For the purpose of demonstration, let's represent the code block to analyze
# This should ideally be read from the current cell's content.
code_block_to_analyze = '''import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Call once to configure Bokeh to display plots inline in the notebook.
output_notebook()
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = ["#%02x%02x%02x" % (r, g, 150) for r, g in zip(np.floor(50+2*x).astype(int), np.floor(30+2*y).astype(int))]

p = figure()
p.circle(x, y, radius=radii, fill_color=colors, fill_alpha=0.6, line_color=None)
show(p)'''

lines = code_block_to_analyze.split('\n')

total_significant_lines = 0
lines_for_color_generation = 0

for line in lines:
    stripped_line = line.strip()
    if stripped_line and not stripped_line.startswith('#'):
        total_significant_lines += 1
        if 'colors = [' in stripped_line and 'for r, g in zip' in stripped_line:
            lines_for_color_generation += 1

# Ensure we have at least 1 line for color generation if it's detected as 0 for some reason
if lines_for_color_generation == 0 and 'colors = [' in code_block_to_analyze and 'for r, g in zip' in code_block_to_analyze:
    lines_for_color_generation = 1

# Fallback if dynamic counting fails or produces zero for total significant lines
if total_significant_lines == 0:
    total_significant_lines = 12 # Default to previous manual count
if lines_for_color_generation == 0:
    lines_for_color_generation = 1 # Default to previous manual count

percentage_color_generation = (lines_for_color_generation / total_significant_lines) * 100
percentage_other_code = ((total_significant_lines - lines_for_color_generation) / total_significant_lines) * 100

print(f'\n**סה"כ שורות קוד משמעותיות (לא כולל שורות ריקות והערות):** {total_significant_lines} שורות.')
print(f'\n*   **קוד ליצירת צבעים:** {lines_for_color_generation} שורה.')
print(f'    *   **אחוז:** {percentage_color_generation:.2f}%\n')

print(f'*   **קוד אחר (ייבוא ספריות, הגדרת נתונים, הגדרת והצגת גרף):** {total_significant_lines - lines_for_color_generation} שורות.')
print(f'    *   **אחוז:** {percentage_other_code:.2f}%\n')

print('_יש לזכור ששורה כמו `p.circle` אמנם משתמשת בצבעים, אך תפקידה העיקרי הוא ציור האלמנטים הגרפיים באופן כללי, ולכן נכללת בחלק ה"אחר"._')
```
